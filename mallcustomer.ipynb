# EDA on Mall Customers

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Users\\user\\Downloads\\Mall_Customers.csv")
```

```python
# Gender distribution
df['Gender'].value_counts()
```

```python
# Average Age
df['Age'].mean()
```

```python
# Average Spending Score
df['Spending Score (1-100)'].mean()
```

```python
# Gender count
sns.countplot(x='Gender', data=df)
plt.show()
```

```python
# Age distribution
sns.histplot(df['Age'], bins=10, kde=True)
plt.show()
```

```python
# Spending Score distribution
sns.histplot(df['Spending Score (1-100)'], bins=10, kde=True)
plt.show()
```

```python
# Age vs Spending Score
sns.scatterplot(x='Age', y='Spending Score (1-100)', hue='Gender', data=df)
plt.show()
```
